```python
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame
# Replace 'df' with the name of your actual DataFrame

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
df_standardized = scaler.fit_transform(df)

# Convert the standardized array back to a DataFrame
df_standardized = pd.DataFrame(df_standardized, columns=df.columns)

# Display the standardized DataFrame
print(df_standardized)
```

```python
#Finding optimal no. of clusters
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans

clusters=range(1,10)
meanDistortions=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(df_standardized)
    prediction=model.predict(df_standardized)
    meanDistortions.append(sum(np.min(cdist(df_standardized, model.cluster_centers_, 'euclidean'), axis=1)) / df_standardized
                           .shape[0])


plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
```

```python
#Set the value of k=5
kmeans = KMeans(n_clusters=5, n_init = 15, random_state=2345)

kmeans.fit(df_standardized)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
#Clculate the centroids for the columns to profile
centroid_df = pd.DataFrame(centroids, columns = list(df_standardized) )
centroid_df
```
